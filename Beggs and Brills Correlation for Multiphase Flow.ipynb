```python
# Mixture velocity
import numpy as np
import tkinter as tk
#import math as mt

app = tk.Tk()
#app.iconbitmap("name.ico")
app.configure(bg="light blue")
app.geometry("850x600+100+50")
app.resizable(width=0,height=0)
app.title("Beggs and Brill Correlation - Group 7")


in_frame = tk.Frame(app,width=350,height=330,bg="#E0F3F5")
in_frame.place(relx=0.03,rely=0.03)

tk.Label(in_frame,text="Gas Velocity", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.1,rely=0.1)
tk.Label(in_frame,text="ft/s", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.78,rely=0.1)
Vg_ = tk.Entry(in_frame)
Vg_.place(relx=0.42, rely=0.1)

tk.Label(in_frame,text="Liquid Velocity", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.1,rely=0.2)
tk.Label(in_frame,text="ft/s", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.78,rely=0.2)
Vl_ = tk.Entry(in_frame)
Vl_.place(relx=0.42, rely=0.2)

tk.Label(in_frame,text="Pipe diameter", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.1,rely=0.3)
tk.Label(in_frame,text="ft", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.78,rely=0.3)
d_ = tk.Entry(in_frame)
d_.place(relx=0.42, rely=0.3)

tk.Label(in_frame,text="Liquid Density", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.1,rely=0.4)
tk.Label(in_frame,text="lbm/cu ft", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.78,rely=0.4)
rho_ = tk.Entry(in_frame)
rho_.place(relx=0.42, rely=0.4)

tk.Label(in_frame,text="Gas Density", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.1,rely=0.5)
tk.Label(in_frame,text="lbm/cu ft", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.78,rely=0.5)
rho1_ = tk.Entry(in_frame)
rho1_.place(relx=0.42, rely=0.5)

tk.Label(in_frame,text="Liquid Vel. Num.", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.1,rely=0.6)
tk.Label(in_frame,text="", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.78,rely=0.6)
rho2_ = tk.Entry(in_frame)
rho2_.place(relx=0.42, rely=0.6)

tk.Label(in_frame,text="Gas viscosity", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.1,rely=0.7)
tk.Label(in_frame,text="Cp", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.78,rely=0.7)
rho3_ = tk.Entry(in_frame)
rho3_.place(relx=0.42, rely=0.7)

tk.Label(in_frame,text="Liquid Viscosity", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.1,rely=0.8)
tk.Label(in_frame,text="Cp", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.78,rely=0.8)
It_ = tk.Entry(in_frame)
It_.place(relx=0.42, rely=0.8)

tk.Label(in_frame,text="Angle", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.1,rely=0.9)
tk.Label(in_frame,text="degrees", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.78,rely=0.9)
tt_ = tk.Entry(in_frame)
tt_.place(relx=0.42, rely=0.9)

tk.Label(in_frame,text="Surface Tension", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.1,rely=0.01)
tk.Label(in_frame,text="", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.78,rely=0.01)
st_ = tk.Entry(in_frame)
st_.place(relx=0.42, rely=0.01)

out_frame = tk.LabelFrame(app,width=350,height=550,bg="#E0F3F5")
out_frame.place(relx=0.5,rely=0.03)

tk.Label(out_frame,text="Mixture velocity, Vm", font=("Calibri",10,"bold"), bg="#E0F3F5").place(relx=0.01,rely=0.05)
tk.Label(out_frame,text="ft/s", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.8,rely=0.05)
P_Vm = tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="white", width=14)
P_Vm.place(relx=0.45,rely=0.05)

tk.Label(out_frame,text="No-slip holdup, λns", font=("Calibri",10,"bold"), bg="#E0F3F5").place(relx=0.01,rely=0.1)
tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.8,rely=0.1)
P_Yns = tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="white", width=14)
P_Yns.place(relx=0.45,rely=0.1)

tk.Label(out_frame,text="Nfr", font=("Calibri",10,"bold"), bg="#E0F3F5").place(relx=0.01,rely=0.15)
tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.8,rely=0.15)
P_Nfr = tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="white", width=14)
P_Nfr.place(relx=0.45,rely=0.15)

tk.Label(out_frame,text="Horizon_liquid_holdup", font=("Calibri",10,"bold"), bg="#E0F3F5").place(relx=0.01,rely=0.2)
tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.8,rely=0.2)
P_Yo = tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="white", width=14)
P_Yo.place(relx=0.45,rely=0.2)

tk.Label(out_frame,text="Correct_factor_Coffit ,C", font=("Calibri",10,"bold"), bg="#E0F3F5").place(relx=0.01,rely=0.25)
tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.8,rely=0.25)
P_C = tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="white", width=14)
P_C.place(relx=0.45,rely=0.25)

tk.Label(out_frame,text="Inclin_Correction_Factor", font=("Calibri",10,"bold"), bg="#E0F3F5").place(relx=0.01,rely=0.3)
tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.8,rely=0.3)
P_phi = tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="white", width=14)
P_phi.place(relx=0.45,rely=0.3)

tk.Label(out_frame,text="Frictional factor ratio", font=("Calibri",10,"bold"), bg="#E0F3F5").place(relx=0.01,rely=0.35)
tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.8,rely=0.35)
P_ff = tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="white", width=14)
P_ff.place(relx=0.45,rely=0.35)

tk.Label(out_frame,text="Reynold's number", font=("Calibri",10,"bold"), bg="#E0F3F5").place(relx=0.01,rely=0.4)
tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.8,rely=0.4)
P_Nre = tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="white", width=14)
P_Nre.place(relx=0.45,rely=0.4)

tk.Label(out_frame,text="No slip friction factor", font=("Calibri",10,"bold"), bg="#E0F3F5").place(relx=0.01,rely=0.45)
tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.8,rely=0.45)
P_fn = tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="white", width=14)
P_fn.place(relx=0.45,rely=0.45)

tk.Label(out_frame,text="Two phase friction factor", font=("Calibri",10,"bold"), bg="#E0F3F5").place(relx=0.01,rely=0.5)
tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.8,rely=0.5)
P_ft = tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="white", width=14)
P_ft.place(relx=0.45,rely=0.5)

tk.Label(out_frame,text="Fric_pres_drop, (dp/dZ_f)", font=("Calibri",10,"bold"), bg="#E0F3F5").place(relx=0.01,rely=0.55)
tk.Label(out_frame,text="psi/ft", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.8,rely=0.55)
P_df = tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="white", width=14)
P_df.place(relx=0.45,rely=0.55)

tk.Label(out_frame,text="Pressure_drop (dp/dZ_e)", font=("Calibri",10,"bold"), bg="#E0F3F5").place(relx=0.01,rely=0.6)
tk.Label(out_frame,text="psi/ft", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.8,rely=0.6)
P_de = tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="white", width=14)
P_de.place(relx=0.45,rely=0.6)

tk.Label(out_frame,text="Total Pressure drop, dp/dZ", font=("Calibri",10,"bold"), bg="#E0F3F5").place(relx=0.01,rely=0.7)
tk.Label(out_frame,text="psi/ft", font=("Calibri",11,"bold"), bg="#E0F3F5").place(relx=0.8,rely=0.7)
P_drop = tk.Label(out_frame,text="", font=("Calibri",11,"bold"), bg="white", width=14)
P_drop.place(relx=0.45,rely=0.7)

def Action():
    # define function for doing the necessary calculations
    Vsg = float(Vg_.get())
    Vsl = float(Vl_.get())
    Phog = float(rho1_.get())
    Nlv = float(rho2_.get())
    muig = float(rho3_.get())
    PhoL = float(rho_.get())
    muiL = float(It_.get())
    θ = float(tt_.get())
    d = float(d_.get())
    oL=float(st_.get())
    g = 32.2
    pattern = None

    print('')
    Vm = Vsl + Vsg
    print('Mixture velocity, Vm =', Vm)
    print('')

    # No-slip holdup
    Yns = Vsl / Vm
    print('No-slip hold-up, lamda ns =', Yns)
    print('')

    # Froude number, Nfr
    Nfr = Vm**2 / (g *d)
    print('Nfr =', Nfr)
    print('')

    # Liquid velocity number, Nlv
    if Nlv != 0:
        Nlv == Nlv

    else:
#         oL = float(input('Liquid Surface Tension: '))
        Nlv = Vsl * ( PhoL / (g * oL) ) ** 0.25
        print('Nlv = ', Nlv)
        print('')

    def get_flowpattern(Yns, Nfr):
        # Calculating to find flow pattern;flow is horizontal
        L1 = 316 * Yns **0.302
        L2 = 0.0009252* Yns**-2.4684
        L3 = 0.10 * Yns ** -1.4516
        L4 = 0.5 * Yns **-6.738

        if (Yns < 0.01 and Nfr < L1) or (Yns >= 0.01 and Nfr < L2):
            fpara = [0.98,0.4846,0.08680]
            typeq = 1
            print('')
            print('Flow Regime is Segregated')
            print('')

        elif Yns >= 0.01 and Nfr <= L2:
            typeq = 2
            fpara = []
            print('')
            print('Flow Regime is Transition')
            print('')

        elif (0.01 <= Yns <0.4 and L3 < Nfr <= L1) or (Yns >= 0.04 and L3 < Nfr <= L1):
            typeq = 3
            fpara = [0.845,0.5351,0.0173]
            print('')
            print('Flow Regime is Intermittent')
            print('')

        elif (Yns<0.4 and Nfr >=L1 or Yns >= 0.4 and Nfr > L4):
            # Constants for calculating liquid hold up goiven a flow pattern
            typeq = 4
            fpara = [1.065,0.5824,0.0609]
            print('')
            print('Flow Regime is Distributed')
            print('')

        else:
            print('Flow Regime not found')


        return typeq, fpara

    typeq, fpara = get_flowpattern(Yns, Nfr)


    ## Horizontal hold up, lambda o\
    if typeq != 2:
        def get_horizonholdup(Yns, Nfr, typeq, fpara):
            a, b, c = fpara
            Yo = (a * Yns**b) / Nfr**c
            print(f'The horizontal liquid hold up = {Yo}')
            return Yo

        Yo = get_horizonholdup(Yns, Nfr, typeq, fpara)

        def get_correct_const(Yns, Nlv, Nfr, typeq):

    #         θ = float(input('Angle of deviation from the horizontal - 90 to 90 , θ = '))

            if 0 < θ <= 90:
                incli = 1
                if typeq == 1:
                    flowpart = 1
                elif typeq == 3:
                    flowpart = 2
                else:
                    flowpart = 3

                C = 0  # Initialize C before the if-elif-else block

                if flowpart == 1:
                    d, e, f, g = [0.011, -3.768, 3.539, -1.614]
                elif flowpart == 2:
                    d, e, f, g = [2.96, 0.305, -0.4473, 0.0978]
                elif flowpart == 3:
                    print('No correction for Transition Flow')
                    C = 0

                if flowpart != 3:
                    C = (1 - Yns) * np.log(d * Yns**e * Nlv**f * Nfr**g)

            else:
                incli = 2
                d,e,f,g = [4.70, -0.3692, 0.1244, -0.5056]
                C = (1 - Yns) * np.log(d * Yns**e * Nlv**f * Nfr**g)

            return C, θ, incli

        C, θ, incli = get_correct_const(Yns, Nlv, Nfr, typeq)
        print('')
        print('Correction factor Cofficient,C = ', C)
        print('')

        ## Correction factor using liquid holdup inclination correction factor
        if θ != 90:

            phi = 1 + C *(np.sin(1.8*θ) - 0.333* np.sin(1.8*θ)**3)
            print('')
            print('Inclination Correction Factor = ', phi)
            print('')

        else:
            phi = 1+0.3*C

        ### Calculate the liquid holdup
        Y = Yo * phi
        print('λ after liquid holdup inclination correction = ', Y)

        if incli == 1:
            #That's uphill
            λ = 0.918 * Y
            print('Applying Palmer Correction Factor for Uphill flow, λ = ', λ)
        else:
            # That's downhil
            λ = 0.541 * Y
            print('Applying Palmer Correction Factor for Downhill flow, λ = ', λ)

    else:
        a1, b1, c1 = [0.98, 0.4846, 0.08680]
        Y1 = (a1 * Yns**b1) / Nfr**c1
        a2, b2, c2 = [0.845, 0.5351, 0.0173]
        Y2 = (a2 * Yns**b2) / Nfr**c2
        L2 = 0.0009252 * Yns**-2.4684
        L3 = 0.10 * Yns ** -1.4516
        a = (L3 - Nfr) / (L3 - L2)
        λ = a * Y1 + (1 - a) * Y2
        #That's transition flow
        print('Liquid holdup λ = ', Y)

    ## Frictional factor ratio
    y = Yns/ λ**2
    if 1 < y < 1.2:
        S = np.log(2.2*y - 1.2)
    else:
        S = np.log(y)/(-0.0523 + 3.182*np.log(y) - 0.8725 * np.log(y)**2 + 0.01853*np.log(y)**4 )

    ffratio = np.e**S

    ##Friction Factor Down
    Phon = PhoL * λ + Phog * (1- λ)
    muin = muiL * λ + muig * (1- λ)
    Nre = 1844 * (Phon * Vm *d)/muin
    fn = 1 / (2*np.log(Nre / (4.5223*np.log(Nre)- 3.8215)))**2
    ftp = fn * ffratio
    dp_f = (ftp * Phon * Vm**2)/ (2*g*d)
    dp_e = Phon
    dp_t = dp_f + dp_e

    P_Vm.configure(text=str(round(Vm,4)))
    P_Yns.configure(text=str(round(Yns,7)))
    P_Nfr.configure(text=str(round(Nfr,4)))
    P_Yo.configure(text=str(round(Yo,7)))
    P_C.configure(text=str(round(C,4)))
    P_phi.configure(text=str(round(phi,4)))
    P_ff.configure(text=str(round(ffratio,4)))
    P_Nre.configure(text=str(round(Nre,4)))
    P_fn.configure(text=str(round(fn,7)))
    P_ft.configure(text=str(round(ftp,7)))
    P_df.configure(text=str(round(dp_f,4)))
    P_de.configure(text=str(round(dp_e,4)))
    P_drop.configure(text=str(round(dp_t,4)))

tk.Button(app,text="   Calculate   ", font=("calibri",12,"bold"),command=Action, bg="dark slate grey",fg="white").place(relx=0.16,rely=0.87)


app.mainloop()
```
